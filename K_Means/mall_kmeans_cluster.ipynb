```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
x=dataset.iloc[:,[3,4]].values
print(x)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('The number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=60,c='red',label='carefull')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=60,c='blue',label='standard')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=60,c='green',label='target')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=60,c='cyan',label='careless')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=60,c='magenta',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroids')
plt.title('cluster of clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
