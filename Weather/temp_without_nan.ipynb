```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('mumbai_bandra.csv')
shape=df.shape
print(shape)
```

```python
df.head()
```

```python
dataset=df.dropna()
```

```python
dataset.isnull().sum()
```

```python
y=dataset.iloc[:,0:1].values
y
```

```python
x=dataset.iloc[:,1:].values
x
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
X = scale.fit_transform(x)
```

```python
scaler = StandardScaler()
```

```python
Y = scaler.fit_transform(y)
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size=0.9,random_state=0)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
scaled = scale.transform([[54.63,0.24,175.27,122.08]])
print([scaled[0]])
```

```python
temp=reg.predict([scaled[0]])
inversed = scaler.inverse_transform(temp)
print(inversed)
```

```python
x_true = scale.fit_transform(x_test)
```

```python
y_pred=reg.predict(x_true)
y_pred_act = scaler.inverse_transform(y_pred)
print(inversed)
```

```python
y_test_act = scaler.inverse_transform(y_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test_act, y_pred_act)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test_act, y_pred_act, squared=False)
```

```python
mean_squared_error(y_test_act, y_pred_act)
```

```python
%matplotlib inline 
plt.plot(abs(y_test_act-y_pred_act))
plt.show()
```
