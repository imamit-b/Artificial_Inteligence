```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('PM2.5.csv')
shape=df.shape
print(shape)
```

```python
df.head()
```

```python
dataset=df.dropna()
```

```python
dataset.isnull().sum()
```

```python
y=dataset['IHBAS-CPCB']
y.head()
```

```python
x=dataset.drop("IHBAS-CPCB", axis='columns')
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_train)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_pred)
```

```python
y_pred_test=reg.predict(x_test)
y_pred_test
```

```python
r2_score(y_test, y_pred_test)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
scale.fit(x_train)
```

```python
scale.mean_
```

```python
x_train_scaled=scale.transform(x_train)
x_test_scaled=scale.transform(x_test)
```

```python
x_train_scaled
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
reg_scaled=linear_model.LinearRegression()
```

```python
reg_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred_scaled=reg_scaled.predict(x_train_scaled)
y_pred_scaled
```

```python
r2_score(y_train, y_pred_scaled)
```

```python
y_pred_test_scaled=reg_scaled.predict(x_test_scaled)
y_pred_test_scaled
```

```python
r2_score(y_test, y_pred_test_scaled)
```
